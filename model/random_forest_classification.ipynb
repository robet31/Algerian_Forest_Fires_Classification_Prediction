#Ubah Tipe data kolom FWI ke float dan menghapus row yang error (tidak sesuai dengan tipe data)

```python
from logging import exception
errorIndication = []

def tryConvert (row) :
  index = dataset.index.get_loc(row.name)
  try :
    dataset.iloc[index,-2] = float(dataset.iloc[index, -2])

  except Exception:
        errorIndication.append(index)

dataset.apply( tryConvert ,axis=1)

print(errorIndication)

for i in errorIndication :
  dataset.drop(i, inplace = True)
  print(i)
```

#feature engineering untuk menghasilkan kolom week

```python
dataset['days'] = pd.to_datetime(dataset[['year', 'month', 'day']])

minggu_pertama = dataset['days'].dt.to_period('M').apply(lambda x: x.start_time.isocalendar()[1])

dataset['week in month'] = dataset['days'].dt.isocalendar().week - minggu_pertama + 1
```

#3.e | hubungan antar variabel (korelasi)

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
dataset['Classes  '] = le.fit_transform(dataset['Classes  '])
```

```python
corr = dataset.corr()

# Buat heatmap dengan seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
threshold = 0.8
collinear = set()
for i in range(len(corr.columns)):
    for j in range(i):
        if abs(corr.iloc[i, j]) > threshold:
            collinear.add(corr.columns[i])

dataset = dataset.drop(columns=collinear)
```

#Random Forest Classification

```python
# dataset.drop('days', inplace=True, axis=1)
```

```python
dataset.head()
```

```python
indexColumn = []

for index, col in enumerate (dataset.columns) :
  indexColumn.append(index)

print(indexColumn)
```

```python
x = dataset.iloc[:, list(filter(lambda x : x != 10, indexColumn))].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
print(y)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [3, 5, 10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf = RandomForestClassifier(n_estimators=100, random_state=42)

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1)
grid_search.fit(x_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
```
